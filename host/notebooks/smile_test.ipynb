Този документ демонстрира как се работи с Arduino и неговото управление от Python. 

Arduino SMAIL фърмуера включва една ранна версия на командният интерпретатор Cmd и библиотеката Ticker, която стартира таскове през определено време, един модул, с които могат да се четат и записват през Cmd няколко float променливи, които управляват генератор на псевдоданни.  


За комуникация с Arduino от PC се използва серииен канал в случая през USB. На компютъра може да се използва серииен порт терминална програма(примерно Arduino или PlatformIO monitor) през, която да се изпращат команди до Arduino и да се получават техните отговори. За същото, в Python среда може да се изполва и библиотеката Pykiba в която има функция за изпращане на команди и парсване на техният отговор. Предложена е допълнена версия на Cmd с вградена команда help, която листва всички въведени команди. В такъв случай може да се използва PykiDev, които разширява Pykiba, като мапва Arduino командите като Python функции.

### Let use PykiDev. It map all Arduino comands like Python functions ...


```python
from pykiba import Pykiba, PykiDev, search_by_manufacturer,  serial_ports_list
import time
from pprint import pprint


name = "Arduino (www.arduino.cc)"       #Linux name of the Arduino Mega2560 driver

ports_found = serial_ports_list()       #Returns a list with all found serial ports
arduino_port_at_the_momnet = search_by_manufacturer(ports_found, name)                                      
print(f'Computer comport of the arduino is: {arduino_port_at_the_momnet}')  

# Lets start comunication with Pykiba...

ard = PykiDev(port = arduino_port_at_the_momnet, baudrate = 115200) # by default baudrate = 9600

time.sleep(2) #must be a delay 

print(f'Returns the name of the project: {ard.command("hello")}') # all Arduino commands can be reached in disl way  
print(f'Returns the name of the project: {ard.hello()}') # or maped like python commands compare with above ...
print(f'Sets mode of operation of the Arduino. Returns setted mode:  { ard.command("mode",1)}') # in mode 1 the led blinks 2 times
print(f'Sets mode of operation of the Arduino. Returns setted mode:  { ard.mode(2)}') # in mode 1 the led blinks 3 times

#some other commandspprint(f'List with available global frolat parameters:')#
print('Format:')
print('[name, value, min, max, update flag]')
pprint(ard.globs())#

#del ard  # To close comport and free it for next example

```

```python
ard.w1(1)
ard.w2(1)
ard.a1(3)
ard.a2(3)
%matplotlib qt

import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation
#%matplotlib inline
# --- Инициализация ---
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()  # втора ордината (споделен X)

xdata = []
y1data, y2data = [], []

line1, = ax1.plot([], [], 'r-', label='f1')  # червен сигнал
line2, = ax2.plot([], [], 'b-', label='f2')  # син сигнал

ax1.set_xlabel("Време / кадри")
ax1.set_ylabel("Сигнал f1", color='r')
ax2.set_ylabel("Сигнал f2", color='b')

max_points = 100

# --- Анимирана функция ---
def animate(i):
    v1 = ard.command('f1')
    v2 = ard.command('f2')
    xdata.append(i)
    y1data.append(v1)
    y2data.append(v2)
    plt.show()
    # ограничаваме броя на точките
    if len(xdata) > max_points:
        xdata.pop(0)
        y1data.pop(0)
        y2data.pop(0)

    # обновяване на данните
    line1.set_data(range(len(y1data)), y1data)
    line2.set_data(range(len(y2data)), y2data)

    # --- динамично мащабиране ---
    ax1.relim()          # преизчислява границите на оста
    ax1.autoscale_view() # мащабира според новите данни
    ax2.relim()
    ax2.autoscale_view()

    ax1.set_xlim(0, max_points)
    return line1, line2

anim = FuncAnimation(fig, animate, interval=1, blit=False, cache_frame_data=False)
plt.show()
```
